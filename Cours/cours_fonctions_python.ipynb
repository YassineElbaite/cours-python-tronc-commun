## 📘 Partie 2 — Créer sa première fonction

### 🔑 La syntaxe `def`

```python
def nom_de_la_fonction():
    # instructions
```

| Mot | Signification |
|-----|---------------|
| `def` | **def**inir une fonction |
| `nom_de_la_fonction` | Le nom qu'on choisit |
| `()` | Les parenthèses — obligatoires |
| `:` | Les deux points — obligatoires |
| *Indentation* | Les instructions de la fonction (4 espaces) |

````python
# ✅ Ma première fonction !
def dire_bonjour():
    print("Bonjour tout le monde !")
    print("Bienvenue au cours de Python 🐍")


# Appeler la fonction (l'utiliser)
dire_bonjour()
dire_bonjour()
````

## 📙 Partie 4 — Le mot-clé `return`

### 🤔 Quelle est la différence ?

Jusqu'ici, nos fonctions **affichent** des résultats avec `print`.
Mais parfois, on veut que la fonction **renvoie** un résultat qu'on peut utiliser ailleurs.

➡️ C'est le rôle de `return` !

| | `print` | `return` |
|---|---|---|
| **Affiche à l'écran** | ✅ Oui | ❌ Non |
| **Renvoie une valeur** | ❌ Non | ✅ Oui |


````python
# ✅ Exemple concret : Calculer la moyenne de 2 notes
def calculer_moyenne(note1, note2):
    moyenne = (note1 + note2 ) / 2
    return moyenne



print("Moyenne de Marwan :",calculer_moyenne(14,12))

````

## 📕 Partie 5 — Fonctions avec conditions

On peut mettre des **conditions `if/else`** à l'intérieur d'une fonction.
La fonction peut retourner des valeurs différentes selon les cas.

````python
# ✅ Fonction qui vérifie si une note est valide (entre 0 et 20)
def note_valide(note):
    if note >=10:
        print("admis")
    else:
        print("ajourné")

note_valide(9)
````

````python
# ✅ Fonction qui donne la mention selon la moyenne
def mention(moyenne):
    if moyenne >= 16:
        return "Très Bien 🌟"
    elif moyenne >= 14:
        return "Bien 👍"
    elif moyenne >= 12:
        return "Assez Bien 👌"
    elif moyenne >= 10:
        return "Passable ✔️"
    else:
        return "Insuffisant ❌"

# Tests
print("Moyenne 17 ->", mention(17))
print("Moyenne 13 ->", mention(13))
print("Moyenne  8 ->", mention(8))
````

````python
# ✅ Combiner les deux fonctions !
def evaluer_eleve(nom, note1, note2, note3):
    moy = calculer_moyenne(note1, note2, note3)
    men = mention(moy)
    print(f"Élève : {nom} | Moyenne : {moy:.2f} | Mention : {men}")

# Tests
evaluer_eleve("Yassine", 15, 17, 18)
evaluer_eleve("Fatima",  12, 11, 13)
evaluer_eleve("Amine",    8,  7,  9)
````

## 🚀 Partie 7 — Mini Projet Final : La Calculatrice 🧮

Nous allons créer une **mini calculatrice** avec des fonctions !

### Étape 1 : Créer les 4 fonctions de calcul

````python
# ✅ Les 4 fonctions de la calculatrice

def addition(a, b):
    return a + b

def soustraction(a, b):
    return a - b

def multiplication(a, b):
    return a * b

def division(a, b):
    if b == 0:
        return "❌ Erreur : Division par zéro impossible !"
    else:
        return a / b

print("✅ Les 4 fonctions sont prêtes !")
````

### Étape 2 : Créer le menu de la calculatrice

````python
# ✅ Mini Calculatrice complète avec menu

def afficher_menu():
    print("\n" + "="*35)
    print("     🧮 MINI CALCULATRICE PYTHON")
    print("="*35)
    print("  1️⃣  Addition")
    print("  2️⃣  Soustraction")
    print("  3️⃣  Multiplication")
    print("  4️⃣  Division")
    print("  5️⃣  Quitter")
    print("="*35)

def calculatrice():
    continuer = True
    while continuer:
        afficher_menu()
        choix = input("Choisissez une opération (1-5) : ")

        if choix == "5":
            print("\nMerci d'avoir utilisé la calculatrice ! À bientôt 👋")
            continuer = False
        elif choix in ["1", "2", "3", "4"]:
            a = float(input("Entrez le premier nombre  : "))
            b = float(input("Entrez le deuxième nombre : "))

            if choix == "1":
                resultat = addition(a, b)
                print(f"\n✅ {a} + {b} = {resultat}")
            elif choix == "2":
                resultat = soustraction(a, b)
                print(f"\n✅ {a} - {b} = {resultat}")
            elif choix == "3":
                resultat = multiplication(a, b)
                print(f"\n✅ {a} × {b} = {resultat}")
            elif choix == "4":
                resultat = division(a, b)
                print(f"\n✅ {a} ÷ {b} = {resultat}")
        else:
            print("❌ Choix invalide ! Entrez un nombre entre 1 et 5.")

# Lancer la calculatrice !
calculatrice()
````
